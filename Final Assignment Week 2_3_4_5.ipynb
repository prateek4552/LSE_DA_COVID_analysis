# Create a subset with required columns

```python
# Create new dataframe with required columns/ drop unnecessary columns
cas_vac1 = cas_vac.iloc[:, [0,7,8,9,10,11,12,13,14]]

# view dataframe
print(cas_vac1.shape)
cas_vac1
```

# Determining people with 'First Dose' only

```python
# Difference between 'First Dose' and 'Second Dose'
# Create a new column
cas_vac1['first_dose_only'] = cas_vac1['First Dose'] - cas_vac1['Second Dose']

# View Dataframe
cas_vac1
```

# Sorting First Dose difference by region

```python
# Determine difference in dosage across various Province/State
cas_vac1.groupby('Province/State')[['first_dose_only', 'Vaccinated', 'First Dose']]\
.sum().sort_values('first_dose_only', ascending=False)
```

# Sorting First Dose (Percentage) by region

```python
# create a new dataframe 'Eligible'
Eligible = pd.DataFrame(cas_vac1.groupby('Province/State')[['first_dose_only', 'Vaccinated', 'First Dose']]\
.sum().sort_values('first_dose_only', ascending=False))
```

```python
# View the dataframe
Eligible
```

```python
# Determine the percentage of people eligible for second dose (who have received first dose only)
# Add new column 'first_dose_percentage'
Eligible['first_dose_percentage'] = Eligible['first_dose_only']*100 / Eligible['First Dose']

# View the output by region
Eligible['first_dose_percentage']
```

# Determine ratio of people with second dose over first dose

```python
# Add a new column 'vaccinated_percentage'
Eligible['vaccinated_percentage'] = Eligible['Vaccinated']*100 / Eligible['First Dose']

# View output by region
Eligible['vaccinated_percentage']
```

# Twitter Analysis

```python
# Read the file
tweets = pd.read_csv('tweets.csv')

# View the dataframe
print(tweets.columns)
print(tweets.shape)
print(tweets.dtypes)
tweets.head()
```

```python
# Create new dataframe with columns containing text value
tweets_text = tweets[['text']]

# view dataframe
tweets_text
```

```python
# Extract #symbol values from Dataframe
def find_tags(row_string):
    # use a list comprehension to find list items that start with #
    tags = []
    if isinstance(row_string, float):
        return tags
    for x in row_string:
        if x.startswith('#'):
            tags.append(x)
            
    return tags
```

```python
# Create a new column 
tweets_text['split'] = tweets_text['text'].str.split(' ')

# view the Output
tweets_text
```

```python
# Extract #symbol values from new column
tweets_text['tags'] = tweets_text['split'].apply(lambda row : find_tags(row))
```

```python
# View the Dataframe
tweets_text
```

```python
# create a loop to extract all #symbol values
def count_tags(tags_list, tags_dict):
    for tag in tags_list:
        if tag in tags_dict:
            tags_dict[tag] += 1
        else:
            tags_dict[tag] = 1
```

```python
# create a dictionary with #symbols and counts
tags_dict = {}
tweets_text['tags'].apply(lambda row : count_tags(row, tags_dict))

# View the output
print(tags_dict)
```

```python
# Import counter 
from collections import Counter

# Determine Top 30 trending hashtags
tags_counter = Counter(tags_dict)
print(tags_counter.most_common(30))
```

```python
# Convert dictionary to dataframe
tags_count_df = pd.DataFrame.from_records(tags_counter.most_common(), columns=['hashtag','count'])

# View the dataframe
tags_count_df
```

```python
# Create Barplot
fig, ax = plt.subplots(dpi=100)
fig.set_size_inches(18,9)
plt.title('Top 25 Trending Hashtags')

sns.barplot(x='hashtag', y='count', data=tags_count_df.head(25))

# alter the x ticks display
plt.xticks(rotation=45)
plt.xlabel('HashTag')
plt.ylabel('Count')

# save the plot
fig.savefig('Top_Trending_Covid_Hashtags', dpi=72)
```

```python
# Create Barplot
fig, ax = plt.subplots(dpi=100)
fig.set_size_inches(18,9)
plt.title('Top 10 Trending Hashtags')

sns.barplot(x='hashtag', y='count', data=tags_count_df.head(10))

# alter the x ticks display
plt.xticks(rotation=45)
plt.xlabel('HashTag')
plt.ylabel('Count')
fig.savefig('Top_10_Trending_Hashtags', dpi=72)
```
